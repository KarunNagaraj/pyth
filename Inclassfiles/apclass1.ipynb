```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.animation import FuncAnimation
from sklearn import datasets
import plotly.express as px
import pandas as pd
import numpy as np

print("Data Analytics Module Loaded Successfully")
```

```python
iris = datasets.load_iris()
X=iris.data
y=iris.target
labels=iris.target_names

#datacleaning
df=pd.DataFrame(X,columns=iris.feature_names)
df.head()
G=df.describe()
print(df)






```

```python
fig=plt.figure()
ax=fig.add_subplot(111,projection='3d')
for i,label in enumerate(labels):
    ax.scatter(X[y==i,0],
               X[y==i,1],
               X[y==i,2],
               label=label)
    
ax.set_xlabel('SEPAL LENGTH (cm)')
ax.set_ylabel('SEPAL WIDTH (cm)')
ax.set_zlabel('PETAL LENGTH (cm)')
ax.legend()

plt.show()
```

```python
#3d plot bar chart for iris dataset, not given by maam 
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import load_iris

# Load Iris dataset
iris = load_iris()
X = iris.data          # shape (150, 4)
y = iris.target        # shape (150,)

# Compute mean of each feature for each class
means = np.array([X[y == i].mean(axis=0) for i in range(3)])
# shape: (3 classes, 4 features)

# Make bar positions
x, y_pos = np.meshgrid(np.arange(3), np.arange(4))
x = x.flatten()
y_pos = y_pos.flatten()
z = np.zeros_like(x)

dx = np.ones_like(x) * 0.5
dy = np.ones_like(x) * 0.5
dz = means.T.flatten()

# Plot
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
ax.bar3d(x, y_pos, z, dx, dy, dz)

# Axis labels
ax.set_xticks(np.arange(3) + 0.25)
ax.set_xticklabels(iris.target_names)

ax.set_yticks(np.arange(4) + 0.25)
ax.set_yticklabels(iris.feature_names)

ax.set_zlabel("Mean value")

plt.show()
```
