```python

size = 10

# create an empty image
image = np.zeros((size, size))

# fill using index logic
for i in range(size):
    for j in range(size):
        if (i + j) % 2 == 1:
            image[i, j] = 1

plt.imshow(image, cmap="gray")
plt.title("10×10 Checkerboard Image")
plt.axis("off")
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# read image (OpenCV reads in BGR)
img_bgr = cv2.imread("image.png")

# convert BGR to RGB for matplotlib
img = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(10, 5))

# original image
plt.subplot(1, 2, 1)
plt.imshow(img)
plt.title("Original Image")
plt.axis("off")

# brightness adjustment (darkening)
brightness_value = -100
darkened = img.astype(np.int16) + brightness_value
darkened = np.clip(darkened, 0, 255).astype(np.uint8)

# darkened image
plt.subplot(1, 2, 2)
plt.imshow(darkened)
plt.title("Darkened Image")
plt.axis("off")

plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

plt.figure(figsize=(15, 4))

img_bgr = cv2.imread("image.png")
img = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)

negative = 255 - img
bright = cv2.convertScaleAbs(img, alpha=1.2, beta=30)
gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
_, thresh = cv2.threshold(gray, 128, 255, cv2.THRESH_BINARY)

# Original
plt.subplot(1, 5, 1)
plt.imshow(img)
plt.title("Original")
plt.axis("off")

# Negative
plt.subplot(1, 5, 2)
plt.imshow(negative)
plt.title("Negative")
plt.axis("off")

# Brightness & Contrast
plt.subplot(1, 5, 3)
plt.imshow(bright)
plt.title("Brightness & Contrast")
plt.axis("off")

# Grayscale
plt.subplot(1, 5, 4)
plt.imshow(gray, cmap="gray")
plt.title("Grayscale")
plt.axis("off")

# Thresholded
plt.subplot(1, 5, 5)
plt.imshow(thresh, cmap="gray")
plt.title("Thresholded")
plt.axis("off")

plt.show()
```

```python
#types of noise
#convultion

#exercise
#read 128*128 image
#extract/slice 5*5 and display the pixel values
#apply image processing calculations to the slice and display the output alongside original

img_bgr = cv2.imread(r"C:\Users\KARUN NAGARAJ\Downloads\advanced python\pngegg.png")

# Extract a 5x5 slice from the image
slice_5x5 = img_bgr[50:55, 50:55]
print("Slice shape:", slice_5x5.shape)
print(slice_5x5)

# Display the original image and the slice
plt.figure(figsize=(10, 5))

plt.subplot(4, 4, 1)
plt.imshow(cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB))
plt.title("Original Image")
plt.axis("off")

plt.subplot(4, 4, 2)
plt.imshow(cv2.cvtColor(slice_5x5, cv2.COLOR_BGR2RGB))
plt.title("5x5 Slice")
plt.axis("off")

plt.show()

#1. Grayscale conversion
gray_image = cv2.cvtColor(slice_5x5, cv2.COLOR_BGR2GRAY)
print("Grayscale Image Slice:")
gray_slice = gray_image
plt.imshow(gray_slice, cmap='gray')
plt.title("Grayscale Slice")
plt.axis("off")
plt.show()

#2. Brightness adjustment
bright_image = cv2.convertScaleAbs(slice_5x5, alpha=1.2, beta=30)
plt.imshow(cv2.cvtColor(bright_image, cv2.COLOR_BGR2RGB))
plt.title("Brightened Slice")
plt.axis("off")
plt.show()

#3. Image Negative
negative_image = 255 - slice_5x5
plt.imshow(cv2.cvtColor(negative_image, cv2.COLOR_BGR2RGB))
plt.title("Negative Slice")
plt.axis("off")
plt.show()

#4. Thresholding
_,thresh_image = cv2.threshold(gray_slice, 128, 255, cv2.THRESH_BINARY)
plt.imshow(thresh_image, cmap='gray')
plt.title("Thresholded Slice")
plt.axis("off")
plt.show()




```
