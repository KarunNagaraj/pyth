```python
given_text = word_tokenize("I can't allow you to go home early")
stop_words = set(stopwords.words('english'))
filtered_list = []
for word in given_text:
    if word.lower() not in stop_words:
        filtered_list.append(word)
print(filtered_list)
print(given_text)
```

```python
tokenizer=RegexpTokenizer("[\w']+")
given_text = tokenizer.tokenize("I can't allow you to go home early")    
stop_words = set(stopwords.words('english'))
filtered_list = []
for word in given_text:
    if word.lower() not in stop_words:
        filtered_list.append(word)
print(filtered_list)
print(given_text)
```

```python
#stemming of filtered list

from nltk.stem import PorterStemmer
stemmer=PorterStemmer()
stemmed_words = [stemmer.stem(word) for word in filtered_list]
print(stemmed_words)
```

```python
#lemmatization of filtered list
nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer
lemmatizer=WordNetLemmatizer()
lemmatized_words = [lemmatizer.lemmatize(word) for word in filtered_list]
print(given_text)
print(lemmatized_words)
print(stemmed_words)
```

```python
#first clean text, give line plot with most common words(stop words, tokenization, lemmitization, at each step count the words), and then create a word cloud
text = """
Education plays an important role in shaping society; people learn throughout their lives, and they keep learning as the world changes. Students study different subjects, are studying late at night, and often reflect on what they have studied and learned. Teachers teach with patience, are teaching new methods every year, and have taught generations of curious minds. As students learn and keep learning, they grow in confidence; they are growing in knowledge, and many have grown into thoughtful, responsible individuals. Technology supports this journey too—it helps students learn faster, continue learning more efficiently, and stay connected to new ideas. Of course, learning is not always easy; some struggle, some pause, and some try again. Still, a society that learns, grows, and keeps improving becomes stronger over time, because education inspires progress and hope 😊📚
"""

```

```python
# ==============================
# TEXT PREPROCESSING WITH NLTK-LIKE STEPS (OFFLINE SAFE)
# ==============================

import re
import string
import matplotlib.pyplot as plt
from collections import Counter
from wordcloud import WordCloud

# ------------------------------
# Input Text
# ------------------------------
text = """
Education plays an important role in shaping society; people learn throughout their lives, and they keep learning as the world changes. Students study different subjects, are studying late at night, and often reflect on what they have studied and learned. Teachers teach with patience, are teaching new methods every year, and have taught generations of curious minds. As students learn and keep learning, they grow in confidence; they are growing in knowledge, and many have grown into thoughtful, responsible individuals. Technology supports this journey too—it helps students learn faster, continue learning more efficiently, and stay connected to new ideas. Of course, learning is not always easy; some struggle, some pause, and some try again. Still, a society that learns, grows, and keeps improving becomes stronger over time, because education inspires progress and hope 😊📚
"""

# ------------------------------
# STEP 1: Clean Text
# ------------------------------
text = text.lower()
text = text.translate(str.maketrans('', '', string.punctuation))
text = re.sub(r'[^\x00-\x7F]+', '', text)  # remove emojis

# ------------------------------
# STEP 2: Tokenization
# ------------------------------
tokens = text.split()
count_tokenized = len(tokens)

# ------------------------------
# STEP 3: Stopword Removal
# ------------------------------
stopwords = {
    'the','is','in','and','as','are','an','of','to','they','their','have',
    'with','this','that','on','for','it','not','over','too','some','what',
    'again','every','into','more','always','because','a','at','becomes',
    'be','by','or','from','so','throughout','many','often','still'
}

tokens_no_stop = [word for word in tokens if word not in stopwords]
count_no_stop = len(tokens_no_stop)

# ------------------------------
# STEP 4: Simple Lemmatization
# ------------------------------
def simple_lemmatize(word):
    if word.endswith("ing") and len(word) > 4:
        return word[:-3]
    if word.endswith("ed") and len(word) > 3:
        return word[:-2]
    if word.endswith("s") and len(word) > 3:
        return word[:-1]
    return word

lemmatized_tokens = [simple_lemmatize(word) for word in tokens_no_stop]
count_lemmatized = len(lemmatized_tokens)

# ------------------------------
# Word Frequency
# ------------------------------
word_freq = Counter(lemmatized_tokens)
most_common = word_freq.most_common(10)

print("Most Common Words:")
print(most_common)

# ------------------------------
# Line Plot (Word Count at Each Step)
# ------------------------------
steps = ["Tokenized", "Stopwords Removed", "Lemmatized"]
counts = [count_tokenized, count_no_stop, count_lemmatized]

plt.figure()
plt.plot(steps, counts, marker='o')
plt.title("Word Count at Each Preprocessing Step")
plt.xlabel("Processing Step")
plt.ylabel("Word Count")
plt.show()

# ------------------------------
# Word Cloud
# ------------------------------
wordcloud = WordCloud(width=800, height=400).generate(" ".join(lemmatized_tokens))

plt.figure()
plt.imshow(wordcloud)
plt.axis("off")
plt.title("Word Cloud (After Lemmatization)")
plt.show()
```
