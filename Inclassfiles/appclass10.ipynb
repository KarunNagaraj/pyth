```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image=cv2.imread("C:\\Users\\KARUN NAGARAJ\\Downloads\\advanced python\\Inclassfiles\\image.png")
image1=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
```

```python
#resizing the image
resized_image=cv2.resize(image,(400,400))

#translating the image
M=np.float32([[1,0,100],[0,1,50]])
translated_image=cv2.warpAffine(image,M,(image.shape[1],image.shape[0]))
```

```python
#rotation

(h,w)=image.shape[:2]
center=(w//2,h//2)  #center of image
M=cv2.getRotationMatrix2D(center,-45,1.0) #45 degree rotation
rotated_image=cv2.warpAffine(image,M,(w,h))
plt.figure(figsize=(10,10))

#scaling
scaled_image=cv2.resize(image,None,fx=0.1,fy=0.1,interpolation=cv2.INTER_CUBIC)

#cropping
cropped_image=image[50:250,50:250]

#flipped
flipped_image=cv2.flip(image,0) #1 for horizontal,0 for vertical,-1 for both

#affine transformation
pts1=np.float32([[50,50],[200,50],[50,200]])
pts2=np.float32([[10,100],[200,50],[100,250]])
M=cv2.getAffineTransform(pts1,pts2)
perspective_image=cv2.warpAffine(image,M,(w,h))

#displaying images
plt.subplot(231),plt.imshow(cv2.cvtColor(resized_image,cv2.COLOR_BGR2RGB)),plt.title("Resized Image")
plt.subplot(232),plt.imshow(cv2.cvtColor(translated_image,cv2.COLOR_BGR2RGB)),plt.title("Translated Image")
plt.subplot(233),plt.imshow(cv2.cvtColor(rotated_image,cv2.COLOR_BGR2RGB)),plt.title("Rotated Image")
plt.subplot(234),plt.imshow(cv2.cvtColor(scaled_image,cv2.COLOR_BGR2RGB)),plt.title("Scaled Image")
plt.subplot(235),plt.imshow(cv2.cvtColor(cropped_image,cv2.COLOR_BGR2RGB)),plt.title("Cropped Image")
plt.subplot(236),plt.imshow(cv2.cvtColor(flipped_image,cv2.COLOR_BGR2RGB)),plt.title("Flipped Image")

plt.show()
```

```python
plt.figure(figsize=(10,10))

# 1. Negative image
negative_image = 255 - image
plt.subplot(3, 3, 1)
plt.imshow(cv2.cvtColor(negative_image, cv2.COLOR_BGR2RGB))
plt.title("Negative Image")
plt.axis("off")

# 2. Brightness and contrast
bright_contrast_image = cv2.convertScaleAbs(image, alpha=1.5, beta=50)
plt.subplot(3, 3, 2)
plt.imshow(cv2.cvtColor(bright_contrast_image, cv2.COLOR_BGR2RGB))
plt.title("Bright and Contrast Image")
plt.axis("off")

# 3. Grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.subplot(3, 3, 3)
plt.imshow(gray_image, cmap='gray')
plt.title("Gray Image")
plt.axis("off")

# 4. Thresholding
_, thresh_image = cv2.threshold(gray_image, 120, 255, cv2.THRESH_BINARY)
plt.subplot(3, 4, 4)
plt.imshow(thresh_image, cmap='gray')
plt.title("Thresholded Image")
plt.axis("off")


#blurring
blurred_image=cv2.GaussianBlur(image,(15,15),0)
plt.subplot(3,3,5)
plt.imshow(cv2.cvtColor(blurred_image,cv2.COLOR_BGR2RGB))
plt.title("Blurred Image")
plt.axis("off")

#edge detection
edge_image=cv2.Canny(image,100,200)
plt.subplot(3,3,6)
plt.imshow(edge_image,cmap='gray')
plt.title("Edge Detected Image")
plt.axis("off")
plt.tight_layout()
plt.show()
```

```python
def mean_filter(image):
    rows, cols = image.shape
    output= image.copy()
    for i in range(1, rows-1):
        for j in range(1, cols-1):
            neighnbors = image[i-1:i+2, j-1:j+2]
            output[i, j] = np.mean(neighnbors)
    return output

filtered_image = mean_filter(gray_image)
plt.imshow(filtered_image, cmap='gray')
plt.title("Mean Filtered Image")
plt.axis("off")
plt.show()
         
```
