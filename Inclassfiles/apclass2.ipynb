```python
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap

# Set the figure size to make sure all subplots are legible
plt.figure(figsize=(18, 10))

# --- 1. Orthographic (ortho) ---
# A "globe" view, centered here on India/Asia
plt.subplot(2, 3, 1)
m = Basemap(projection='ortho', lat_0=20, lon_0=78, resolution='l')
m.drawcoastlines()
m.fillcontinents(color='#cc9966', lake_color='#99ffff')
m.drawmapboundary(fill_color='#99ffff')
plt.title("Orthographic (ortho)")

# --- 2. Geostationary (geos) ---
# View from a satellite height (default 35,786 km)
plt.subplot(2, 3, 2)
m=Basemap(projection='geos',lon_0=105,resolution='l',rsphere=(6378137.00,6356752.3142))
m.drawcoastlines()
m.fillcontinents(color='#cc9966', lake_color='#99ffff')
m.drawmapboundary(fill_color='#99ffff')
plt.title("Geostationary (geos)")

# --- 3. Hammer (hammer) ---
# Global Equal-Area projection
plt.subplot(2, 3, 3)
m = Basemap(projection='hammer', lon_0=0, resolution='l')
m.drawcoastlines()
m.drawcountries()
m.fillcontinents(color='#cc9966', lake_color='#99ffff')
m.drawmapboundary(fill_color='#99ffff')
plt.title("Hammer (hammer)")

# --- 4. Robinson (robin) ---
# Global compromise projection (good for world maps)
plt.subplot(2, 3, 4)
m = Basemap(projection='robin', lon_0=0, resolution='l')
m.drawcoastlines()
m.drawcountries()
m.fillcontinents(color='#cc9966', lake_color='#99ffff')
m.drawmapboundary(fill_color='#99ffff')
plt.title("Robinson (robin)")

# --- 5. Mercator (merc) ---
# Standard navigation map. Requires bounding box corners.
plt.subplot(2, 3, 5)
m = Basemap(projection='merc', llcrnrlat=-60, urcrnrlat=80,
            llcrnrlon=-180, urcrnrlon=180, lat_ts=20, resolution='c')
m.drawcoastlines()
m.fillcontinents(color='#cc9966', lake_color='#99ffff')
m.drawmapboundary(fill_color='#99ffff')
plt.title("Mercator (merc)")

# --- 6. Lambert Conformal Conic (lcc) ---
# Regional map, good for shapes. Centered on India.
plt.subplot(2, 3, 6)
m = Basemap(projection='lcc', width=4000000, height=4000000,
            lat_0=24, lon_0=78, lat_1=15, lat_2=30, resolution='l')
m.drawcoastlines()
m.drawcountries()
m.fillcontinents(color='#cc9966', lake_color='#99ffff')
m.drawmapboundary(fill_color='#99ffff')
plt.title("Lambert Conformal Conic (lcc)")

# Adjust layout to prevent overlap
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap

# specific setup to make the plot look nice
plt.figure(figsize=(14, 10))
plt.subplots_adjust(hspace=0.3) # add space between rows

# -----------------------------------------------------------
# 1. MERCATOR ('merc')
# -----------------------------------------------------------
plt.subplot(2, 2, 1)
# Global map, cutting off the extreme poles
m1 = Basemap(projection='merc', 
             llcrnrlat=-80, urcrnrlat=80,
             llcrnrlon=-180, urcrnrlon=180,
             resolution='c')

m1.drawcoastlines()
m1.fillcontinents(color='brown', lake_color='white')
m1.drawmapboundary(fill_color='black')
# Draw lat/lon grid lines every 60 degrees
m1.drawparallels(range(-90, 91, 60), labels=[1,0,0,0]) 
m1.drawmeridians(range(-180, 181, 60), labels=[0,0,0,1])
plt.title("Mercator ('merc')\nNote the size of Greenland vs Africa")


# -----------------------------------------------------------
# 2. ORTHOGRAPHIC ('ortho')
# -----------------------------------------------------------
plt.subplot(2, 2, 2)
# Centered on the Atlantic to show Americas and Africa
m2 = Basemap(projection='ortho', lat_0=20, lon_0=-60, resolution='c')

m2.drawcoastlines()
m2.fillcontinents(color='lightgray', lake_color='white')
m2.drawmapboundary(fill_color='white')
m2.drawparallels(range(-90, 91, 30))
m2.drawmeridians(range(-180, 181, 30))
plt.title("Orthographic ('ortho')\nPerspective view from Space")


# -----------------------------------------------------------
# 3. ROBINSON ('robin')
# -----------------------------------------------------------
plt.subplot(2, 2, 3)
# Centered on Greenwich (lon_0=0)
m3 = Basemap(projection='robin', lon_0=0, resolution='c')

m3.drawcoastlines()
m3.fillcontinents(color='lightgray', lake_color='white')
m3.drawmapboundary(fill_color='white')
m3.drawparallels(range(-90, 91, 30), labels=[1,0,0,0])
m3.drawmeridians(range(-180, 181, 60), labels=[0,0,0,1])
plt.title("Robinson ('robin')\nThe 'Textbook' Compromise")


# -----------------------------------------------------------
# 4. LAMBERT CONFORMAL ('lcc')
# -----------------------------------------------------------
plt.subplot(2, 2, 4)
# Focused on North America (Global LCC is rarely used)
# We define width and height in meters
m4 = Basemap(projection='lcc', resolution='c',
             lat_0=45, lon_0=-100,
             width=6000000, height=4000000)

m4.drawcoastlines()
m4.fillcontinents(color='lightgray', lake_color='white')
m4.drawcountries() # Added country borders for context
m4.drawmapboundary(fill_color='white')
plt.title("Lambert Conformal ('lcc')\nRegional (North America Focus)")

plt.show()
```
